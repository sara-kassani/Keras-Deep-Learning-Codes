# Data Shapes

```python
import numpy as np
```

### One Dimensional Data

```python
data = np.random.rand(3)
print (data)
```

### Two Dimensional Data

```python
data= np.random.rand(5,4)
print(data)
```

### Three Dimensional Data

```python
data= np.random.rand(5,4,3)
print(data)
```

### Four Dimensional Data

```python
data= np.random.rand(5,4,3,2)
print(data)
```

## Data Reshaping 

### One Dimensional Data to Single Column Matrix

```python
data= np.random.rand(10)
print(data)
```

```python
data.reshape(10,1)
```

### One Dimensional Data to Multiple Column Matrix

```python
data.reshape(5,2)
```

### Check the Shape of our Data

```python
print(data.shape)
```
